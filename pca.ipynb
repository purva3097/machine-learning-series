# PCA

```python
# PCA - reducing 4 dimensional data into 1 dim
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
from sklearn import datasets
iris = datasets.load_iris()
iris
```

```python
x = pd.DataFrame(iris.data, columns= iris.feature_names)
x
```

```python
# take target column as y
y = iris.target # 0 - satosa, 1 - versicolor, 2 - virginica
y
```

```python
# The idea behind StandardScaler is that it will transform our data such that its distribution
# will have a mean value 0 and standard deviation of 1. (Standard Normal Distribution)
# Given the distribution of the data, each value in the dataset will have the sample mean value 
# subtracted, and then divided by the standard deviation of the whole dataset. 
# Note: We can also use MinMaxScaler here. 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
# Display covariance matrix.
# The columns in the x_scaled will become rows for calculating covariance matrix
rows = x_scaled.T # find transpose of x_scaled
covariance_matrix = np.cov(rows)
print(covariance_matrix)
```

```python
# Eigen values and eigen vectors 
# eigenvectors (principal components) determine the directions of the new feature space
# and the eigenvalues determine their magnitude.
eig_vals, eig_vecs = np.linalg.eig(covariance_matrix)
print('Eigenvectors \n%s' %eig_vecs)
print('\nEigenvalues \n%s' %eig_vals)
```

```python
# let us see which eigen value has highest variance
# we can see eig_vals[0] has 73% of variance (hence information gain)
for i in range(len(eig_vals)):
 print(eig_vals[i] / sum(eig_vals))
```

```python
# project data points on to selected eigen vector
# eig_vecs.T[0] -> gives the 0th col of eigen vectors
projected_x = x_scaled.dot(eig_vecs.T[0])
projected_x
```

```python
# now compose the dataset for our analysis
result = pd.DataFrame(projected_x, columns=['PC1'])
result['y-axis'] = 0.0
result['label'] = y
result
```

```python
# draw a plot with points 
sns.lmplot('PC1', 'y-axis', data=result, fit_reg=False, # x-axis, y-axis, data, no line 
 hue="label") # color
```

```python
# Method 2: using PCA class in sklearn
from sklearn import decomposition
pca = decomposition.PCA(n_components=1) # reduce dim to 1
pca_x = pca.fit_transform(x_scaled)
```

```python
# draw the plot
result1 = pd.DataFrame(pca_x, columns=['PC1'])
result1['y-axis'] = 0.0
result1['label'] = y
sns.lmplot('PC1', 'y-axis', data = result1, fit_reg=False, # x-axis, y-axis, data, no line
 hue="label") # color
```

```python
# display dimensionality
print(x_scaled.shape) # (150, 4)
print(pca_x.shape) # (150, 1)
```

```python
# divide the data into two parts
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(pca_x, y, test_size=0.2)
```

```python
# train the model. We are using Support Vector Machine
from sklearn.svm import SVC 
model = SVC() # C=1.0, kernel='rbf', gamma='auto_deprecated'
model.fit(x_train, y_train) 
```

```python
# find accuracy level
model.score(x_test, y_test)
```

```python
# enter our own data for prediction
inputs = np.array([[6.6,3.0,4.4,1.4]]) 
```

```python
# convert inputs using StandardScaler
scaled = scaler.transform(inputs)
```

```python
# apply pca and transform (do not use fit)
pca_inputs = pca.transform(scaled)
```

```python
# make prediction
model.predict(pca_inputs)
```
