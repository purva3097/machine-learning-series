1)Guassian Naive Bayes Classifier.

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
```

```python
#readinng the csv file
df=pd.read_csv("D:/ds datatechs/datasets/golf.csv")
df.head()
```

```python
#Encoding the strings to numeric
Numerics=LabelEncoder()
```

```python
#Dropping the target variable and make it as a newframe.
inputs=df.drop('PLAY GOLF', axis='columns')
target=df['PLAY GOLF']
target
```

```python
#creating new dataframe
inputs["OUTLOOK_n"]=Numerics.fit_transform(inputs['OUTLOOK'])
inputs["TEMPERATURE_n"]=Numerics.fit_transform(inputs['TEMPERATURE'])
inputs["HUMIDITY_n"]=Numerics.fit_transform(inputs['HUMIDITY'])
inputs["WINDY_n"]=Numerics.fit_transform(inputs['WINDY'])
inputs
```

```python
#Dropping string values 
inputs_n=inputs.drop(['OUTLOOK','TEMPERATURE','HUMIDITY','WINDY'], axis='columns')
inputs_n
```

```python
#Applying Gaussian Naive Bayes
classifier=GaussianNB()
classifier.fit(inputs_n,target)
```

```python
#Check accuraccy
classifier.score(inputs_n,target)
```

85% accuraccy

```python
#Prediction
classifier.predict([[1,0,1,0]])
```

This means the at this condition weather is suitable for playing golf.
