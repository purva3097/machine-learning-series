# Lasso Regression

TASK:-Using Lasso regression on boston house prices, select the best feature that mostly influences the output(prices).

```python
#code for feature selection in Lasso regression
import pandas as pd
import matplotlib.pyplot as plt
```

```python
boston=pd.read_csv("D:/ds datatechs/datasets/boston_houses.csv")
boston.head()
```

```python
#delete the price column as it is dependent and take it as y
x=boston.drop('Price', axis=1)
x
```

```python
y=boston['Price'].values
y
```

```python
#take the names of columns
names=x.columns
names
```

```python
#rage of columns
rng=range(len(names))
rng
```

```python
#draw line plot between range and coefficients in lasso formula
from sklearn.linear_model import Lasso
ls=Lasso(alpha=0.1)
ls_coef=ls.fit(x,y).coef_
ls_coef
```

```python
plt.plot(rng,ls_coef)
plt.xticks(rng,names,rotation=60)
plt.ylabel("coefficients")
plt.show()
```

From the above graph, we can understand that RM(no. of rooms) is highly influencing the result of (Price).so,this feature should be given highest importance in predicting price.

```python
#Prediction using Lasso regression-we are extending the program
#split data into train and test parts and then train the model
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.3, random_state=0)
ls.fit(x_train,y_train)
```

```python
#predict y vlue based on test data
y_pred=ls.predict(x_test)
```

```python
#find the score of lasso model
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
